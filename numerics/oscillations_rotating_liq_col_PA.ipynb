### Oscillations of a rotating liquid column - Pratik Aghor

```python
# redfine beta = i gamma #
# define alpha as a function of scaled gamma = gamma/Omega and wavenumber a = kR_{2}#
def alpha(a, beta):
    I = CC.0 
    return I*a*sqrt(1 - 4.0/(beta*beta))
```

```python
# define f(gamma), whose zeros give the dispersion relation #
# redfine beta = i gamma #
def f(m, a, beta): 
    # only a fn of beta #
    I = CC.0
    alph = alpha(a, beta) 
    return beta + 0.5*alph*beta*(bessel_J(m-1, alph) - bessel_J(m+1, alph) ) + 2*m*bessel_J(m, alph)
```
