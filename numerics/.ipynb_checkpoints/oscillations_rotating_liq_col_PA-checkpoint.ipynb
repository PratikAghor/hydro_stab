### Oscillations of a rotating liquid column - Pratik Aghor

```python
# redfine beta = i gamma #
# define alpha as a function of scaled gamma = gamma/Omega and wavenumber a = kR_{2}#
def alpha(a, beta):
    I = CC.0 
    return I*a*sqrt(1 - 4.0/(beta*beta))
```

```python
# define f(gamma), whose zeros give the dispersion relation #
# redfine beta = i gamma #
def f(m, a, beta): 
    # only a fn of beta #
    I = CC.0
    alph = alpha(a, beta) 
    return beta + 0.5*alph*beta*(bessel_J(m-1, alph) - bessel_J(m+1, alph) ) + 2*m*bessel_J(m, alph)
```

```python
import numpy as np
m = 1; Nr = 10; a = 10;
var('beta') # define beta to be a variable #
```

```python
plot(f(m, a, beta), beta, (beta, -2, 2)) # [-2, 2] as from the dispersion relation, beta = i gamma varies in there
```

```python
import numpy as np
n = 10000
nprint = n/10
a_vec = np.linspace(10, 0, n) # vector of wavenumbers a = kR_{2} #
root_vec = np.zeros(len(a_vec))
data_1 = np.zeros((len(a_vec), 2))
data_2 = np.zeros((len(a_vec), 2))
data_3 = np.zeros((len(a_vec), 2))
data_4 = np.zeros((len(a_vec), 2))
```

```python
known_roots = np.array([1.7090202082413584, 1.54355099989339, -1.73928019408651, -1.56791082279286]) # known roots at a = 10 #
for j in range(0, len(known_roots)):
    x0 = known_roots[j]
    for i in range(0, n):
        a = a_vec[i]    
        x1 = findroot(Freal, x0) # find root #
        root_vec[i] = x1 # store root #
        data_1[i, 0] = a; data_1[i, 1] = x1;
        if(i % int(nprint) == 0):
            print 'a = ', a, 'x0 = ', x0
        x0 = x1 # update #
```
